```python
import pandas as pd
```

```python
#Create DataFrame from dict of equal length lists
#last five olymics data: place, year and number of countries particpated
olympic_data_list = {'HostCity':['London','Seattle','Athens','Sydney','Atlanta'],
                    'Year':[2012,2008,2004,2000,1992],
                    'No. of Participating Countries':[205,204,201,200,194]}
```

```python
df_olympic_data = pd.DataFrame(olympic_data_list)
```

```python
df_olympic_data
```

```python
#view a DataFrame by referring to the column name or with the sescribe function
#select by City name
df_olympic_data.HostCity
```

```python
#use describe function to display the content
df_olympic_data.describe
```

```python
#Create DataFrame from dict of Series
olympic_series_participation = pd.Series([205,204,201,200,194],index=[2012,2008,2004,2000,1992])
olympic_series_country = pd.Series (['London','Seattle','Athens','Sydney','Atlanta'],
                                    index=[2012,2008,2004,2000,1992])
         
```

```python
df_olympic_series = pd.DataFrame({'No. of Participating Countries':olympic_series_participation,
                                 'Host Cities':olympic_series_country})
```
