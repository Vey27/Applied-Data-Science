```python
# Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To view the plot in notebook
%matplotlib inline
```

```python
#imort the auto dataset
df = pd.read_csv('auto_data_mpg.csv')
```

```python
#view the top 5 record
df.head(5)
```

```python
#write a user defined function for origin
# 1-USA, 2-Europe, 3-Asia
def origin(num):
  if num==1:
    return 'USA'
  elif num==2:
    return 'Europe'
  else:
    return 'Asia'

#use apply function
df['origin'] = df['origin'].apply(origin)
```

```python
#view first 30 data points after applying the user defined function to dataset
df.head(30)
```

```python
#draw the pair plot using sns for mpg, weight, orign, and the hue origin, set the size to 4
#note: hus is variable in data set to mapplot aspects to different colors
sns.pairplot(df[['mpg','weight','origin']],hue='origin',size=4)
```
