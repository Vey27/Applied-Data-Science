```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.csv',parse_dates=True)
df.head()
```

```python
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
```

```python
# Plot the 3D graph
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

```python
# Plot the 3D graph
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['Volume'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('Volume')
plt.show()
```

There is a correlation between the 'High' and 'Low' variables.

```python
def return_z(x,y):
  return 50-(x**2+y**2)
```

```python
sns.set_style('whitegrid')
```

```python
x1,y1 = np.linspace(-5,5,50),np.linspace(-5,5,50)
```

```python
x1,y1 = np.meshgrid(x1,y1)
z1 = return_z(x1,y1)
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,y1,z1)
plt.show()
```
