```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.csv',parse_dates=True)
df.head()
```

```python
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
```

```python
# Plot the 3D graph
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

```python
# Plot the 3D graph
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['Volume'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('Volume')
plt.show()
```

There is a correlation between the 'High' and 'Low' variables.
