#Performing classification


```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Split the data into features (X) and target (y)
X = iris.data[:, [2, 3]]  # Selecting petal length and petal width columns
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

#Training predictions


```python
training_prediction = log_reg.predict(X_train)
training_prediction
```

#Test predictions


```python
test_prediction = log_reg.predict(X_test)
test_prediction
```

#Performance Measures

```python
from sklearn import metrics

print("Precision, Recall, Confusion matrix, in training\n")

# Precision Recall scores
print(metrics.classification_report(y_train, training_prediction, digits=3))

# Confusion matrix
print(metrics.confusion_matrix(y_train, training_prediction))
```

The classification report provides an evaluation of the model's performance on the training data. It includes metrics such as precision, recall, and F1-score for each class, as well as the accuracy of the model.

Precision measures the proportion of correctly predicted instances of a particular class out of all instances predicted as that class. Recall, also known as sensitivity or true positive rate, measures the proportion of correctly predicted instances of a particular class out of all instances that belong to that class in the ground truth. The F1-score is the harmonic mean of precision and recall, providing a single metric that balances both measures.

Looking at the classification report:

- Class 0: It has perfect precision, recall, and F1-score of 1.0. This suggests that all instances of class 0 were correctly predicted, and there were no false positives or false negatives.

- Class 1: It has a precision of 0.923, which means that 92.3% of the instances predicted as class 1 were correct. The recall is 0.889, indicating that 88.9% of the actual instances of class 1 were correctly predicted. The F1-score is 0.906, which balances precision and recall.

- Class 2: It has a precision of 0.893, indicating that 89.3% of the instances predicted as class 2 were correct. The recall is 0.926, suggesting that 92.6% of the actual instances of class 2 were correctly predicted. The F1-score is 0.909, which balances precision and recall.

The accuracy of the model on the training data is 0.933, meaning that it correctly classified 93.3% of the instances.

The confusion matrix provides a tabular representation of the model's predictions compared to the actual values. It shows the counts of true negatives (top-left), false positives (top-right), false negatives (bottom-left), and true positives (bottom-right) for each class. In this case, the confusion matrix indicates that class 0 had 21 true positives, class 1 had 24 true positives, 3 false positives, and class 2 had 25 true positives and 2 false positives.
